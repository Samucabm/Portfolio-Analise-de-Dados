# Goals Graphic
 Create a bar graphic showing the sales of 4 salesmen and a horizontal line representing the sales goal.


```python
# Importing
import matplotlib.pyplot as plt 
```

```python
# Creation of the simulated data in lists
sales = [20, 12, 30, 17]
goal = [25, 12, 33, 22]
sellers = ['Seller 1', 'Seller 2', 'Seller 3', 'Seller 4']
```

```python
# Ploting the data
plt.figure(figsize=(8, 4))
plt.plot(sellers, goal)
plt.bar(sellers, sales)
plt.title('Goals per sales')
plt.ylabel('Sales')
plt.xlabel('Sellers')
plt.show()
```
