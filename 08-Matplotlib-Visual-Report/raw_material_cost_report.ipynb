Create a line graphic with Matplotlib showing the price evolution of 2-3 raw materials throughout the year. Use subtitles, titles and notes to highlight the high and low peaks.

```python
# Imports and Data
import matplotlib.pyplot as plt

# It's a better practice to structure related data in a dictionary
# This makes the code cleaner and easier to loop through
costs = {
    'Steel': [5000, 3000, 4000, 2000, 1000],
    'Copper': [6000, 4000, 5000, 3000, 2000],
    'Aluminum': [4500, 3500, 3000, 2500, 1500]
}
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
```

```python
# Plotting the Graph
# Create the figure
plt.figure(figsize=(12, 7))

# --- Dynamic Plotting with a Loop ---
# This loop iterates through our dictionary, plotting each material automatically
for material, material_costs in costs.items():
    plt.plot(months, material_costs, label=f'{material} Costs', marker='o') # marker='o' adds circles on data points

# --- Customization and Annotations (The Key Part) ---
# Add title and labels
plt.title('Monthly Raw Material Cost Variation')
plt.ylabel('Cost (R$)')
plt.xlabel('Month')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6) # Add a grid for better readability

# --- Adding Annotations to Highlight Peaks and Lows ---
# Let's highlight the peak cost of Copper and the lowest cost of Steel
# plt.annotate() is the tool for writing text on the chart

# Highlight Copper's peak
peak_copper_cost = max(costs['Copper'])
peak_copper_month = months[costs['Copper'].index(peak_copper_cost)]
plt.annotate('Copper Peak', 
             xy=(peak_copper_month, peak_copper_cost), # The point (x,y) to annotate
             xytext=('Feb', 6000),                     # The position (x,y) of the text
             arrowprops=dict(facecolor='black', shrink=0.05))

# Highlight Steel's lowest point
low_steel_cost = min(costs['Steel'])
low_steel_month = months[costs['Steel'].index(low_steel_cost)]
plt.annotate('Steel Lowest Point',
             xy=(low_steel_month, low_steel_cost),
             xytext=('Apr', 1000),
             arrowprops=dict(facecolor='black', shrink=0.10))

# Show the final, annotated plot
plt.show()
```
