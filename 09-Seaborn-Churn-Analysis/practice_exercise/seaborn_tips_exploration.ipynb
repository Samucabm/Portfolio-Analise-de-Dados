```python
# Imports and Data Loading
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example dataset from Seaborn
tips_df = sns.load_dataset('tips')
```

## 1. How many customers are there per day of the week?

```python
# Countplot

plt.figure(figsize=(8, 5))

# Create the count plot. We can get the order directly from the data's categories.
sns.countplot(data=tips_df, x='day', order=['Thur', 'Fri', 'Sat', 'Sun'], palette='viridis')

# Add professional titles and labels
plt.title('Customer Count by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Customers')

plt.show()
```

## 2. What is the distribution of the total bill between lunch and dinner?

```python
# Boxplot
plt.figure(figsize=(8, 6))

# Create the box plot to compare the numerical 'total_bill' across the categorical 'time'
sns.boxplot(data=tips_df, x='time', y='total_bill', palette='pastel')

# Add professional titles and labels
plt.title('Distribution of Total Bill for Lunch vs. Dinner')
plt.xlabel('Meal Time')
plt.ylabel('Total Bill (USD)')

plt.show()
```
