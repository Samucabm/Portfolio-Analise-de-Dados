# Project 9: Risk Factor Analysis for Customer Churn
## Objective:
### To use Seaborn's statistical plots to investigate key hypotheses about customer churn in a telecommunications company.

```python
# Imports and Data Loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## 1. Do customers with month-to-month contracts churn more?
### We will use a `countplot` with the `hue` parameter to compare the churn count within each contract type.

```python
# Countplot for Contract vs. Churn
plt.figure(figsize=(10, 6))

sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')

plt.title('Churn Count by Contract Type', fontsize=16)
plt.xlabel('Contract Type', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(title='Churn', labels=['No', 'Yes'])

plt.show()
```

### **Insight:** The chart clearly shows that the vast majority of churn comes from customers with Month-to-month contracts. This is a critical business insight.

## 2. Is the monthly charge distribution different for customers who churn?
### A `boxplot` is the perfect tool to compare the distribution of a numerical variable (`MonthlyCharges`) across two categories (`Churn`).

```python
# Boxplot for Monthly Charges vs. Churn
plt.figure(figsize=(8, 7))

sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Set2')

plt.title('Distribution of Monthly Charges by Churn Status', fontsize=16)
plt.xlabel('Churn Status', fontsize=12)
plt.ylabel('Monthly Charges (USD)', fontsize=12)
# The plt.xticks line is not necessary as Seaborn handles it automatically.

plt.show()
```

### **Insight:** The boxplot reveals that customers who churn tend to have a higher median monthly charge. Additionally, the range of charges for churning customers is much wider, indicating more variability.
